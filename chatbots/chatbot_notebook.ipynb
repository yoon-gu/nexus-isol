```python
import gradio as gr
import random
import time
import mlflow

# MLflow ì„¤ì • (ì„ íƒì‚¬í•­)
mlflow.set_tracking_uri("http://127.0.0.1:5000")
mlflow.set_experiment("gradio-chatbot-notebook")

# ê°„ë‹¨í•œ ì±—ë´‡ ì‘ë‹µ í•¨ìˆ˜
@mlflow.trace
def respond(message, chat_history):
    # ê°„ë‹¨í•œ ì‘ë‹µ ë³€í˜•
    responses = [
        f"You said: {message}",
        f"I heard: {message}",
        f"Interesting! You mentioned: {message}",
        f"Thanks for sharing: {message}",
        f"I'm a simple chatbot. You told me: {message}"
    ]
    
    # ì²˜ë¦¬ ì‹œê°„ ì‹œë®¬ë ˆì´ì…˜
    time.sleep(1)
    
    # ëœë¤ ì‘ë‹µ ì„ íƒ
    bot_message = random.choice(responses)
    chat_history.append((message, bot_message))
    
    return "", chat_history
```

```python
# Gradio ì¸í„°í˜ì´ìŠ¤ ìƒì„±
with gr.Blocks() as demo:
    gr.Markdown("# ğŸ¤– ì£¼í”¼í„° ë…¸íŠ¸ë¶ ì±—ë´‡")
    
    # ì±—ë´‡ ì¸í„°í˜ì´ìŠ¤
    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="ë©”ì‹œì§€ë¥¼ ì…ë ¥í•˜ì„¸ìš”")
    clear = gr.Button("ëŒ€í™” ì´ˆê¸°í™”")
    
    # ë©”ì‹œì§€ í•¸ë“¤ëŸ¬ ì„¤ì •
    msg.submit(respond, [msg, chatbot], [msg, chatbot])
    
    # ì´ˆê¸°í™” ë²„íŠ¼ ê¸°ëŠ¥
    def clear_chat():
        return None, []
    
    clear.click(clear_chat, None, [msg, chatbot], queue=False)

# ë…¸íŠ¸ë¶ì—ì„œ ì‹¤í–‰
demo.launch(debug=True, share=True)
```
